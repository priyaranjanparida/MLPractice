```python
# Make the Label Value an int data type 
data['y'] = data['y'].replace({'no':0,'yes':1})
```

```python
# To see all the the columns after 
pd.set_option('display.max_columns', None)
# One-hot encoding to convert categorical varibales into numerical variables 
data = pd.get_dummies(data)
```

```python
data.sample(5)
```

```python
# Define input and output columns of the data set 
label='y'
feature = data.columns
feature = feature.drop(['y'])
X= data[feature]
y= data[label]
```

```python
# Import required models for evaluation 
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# Train-Test Splitting of database
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=111)
```

```python
print ("x_train :" , X_train.shape )
print ("x_test :", X_test.shape)
print ("y_train :", y_train.shape)
print ("y_test :", y_test.shape)
```

```python
models = []
models.append(('LR', LogisticRegression(max_iter=500)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier()))
models.append(('SVM', SVC()))
```

```python
models
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from pprint import pprint
```

```python
results = []
names = []
scoring_metrics = ['accuracy', 'f1']
```

```python
y2 = np.ravel(y)
for name, model in models:
    print ('\n========= Model Name is ', name , ' =========')
    kfold = KFold(n_splits=10)
    cv_results = cross_validate(model, X, y2, cv=5, scoring=scoring_metrics, n_jobs=-1)
    pprint (cv_results)
    this_result = cv_results['test_accuracy']
    results.append(this_result)
    names.append(name)
    print("\n{}: accuracy : avg={:,.2f}, median={:,.2f},  std={:,.2f},  cv_results={}".format (
        name, np.mean(this_result), np.median(this_result), np.std(this_result), this_result))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize = (10,6))

# boxplot algorithm comparison
ax = sns.boxplot(data=results)
ax.set_title('Algorithm Comparison')
ax.set_xticklabels(names)
ax.set_ylabel('accuracy')
plt.show()
```
